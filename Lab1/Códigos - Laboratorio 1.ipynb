```python
data1 = loadtxt("Datos1.txt", float)
```

```python
data2 = loadtxt("Datos2.txt", float)
```

```python
data3 = loadtxt("Datos3.txt", float)
```

Datos1 tiene 36 celdas buenas.

Datos2 tiene 35 celdas buenas y 1 semimala.

Datos3 tiene 35 celdas buenas y 1 celda mala.

```python
V1 = data1[:, 0]
I1 = data1[:, 1]
d1 = len(V1)
P1 = empty(shape = d1)
for i in range(d1):
    P1[i] = (data1[i, 0])*(data1[i, 1])
MP1 = max(P1)
print(" El pico de la primera data es: ", MP1)
############################
V2 = data2[:, 0]
I2 = data2[:, 1]
d2 = len(V2)
P2 = empty(shape = d2)
for i in range(d2):
    P2[i] = (data2[i, 0])*(data2[i, 1])
MP2 = max(P2)
print(" El pico de la primera data es: ", MP2)
############################
V3 = data2[:, 0]
I3 = data2[:, 1]
d3 = len(V3)
P3 = empty(shape = d3)
for i in range(d3):
    P3[i] = (data3[i, 0])*(data3[i, 1])
MP3 = max(P3)
print(" El pico de la primera data es: ", MP3)
###########################
plot(V1, P1, label = '36 celdas buenas')
plot(V2, P2, label = '35 celdas buenas y 1 semimala')
plot(V3, P3, label = '35 celdas buenas y 1 mala')
legend(loc = 'best')
title('Curva Potencia(P) vs Voltaje(V)')
grid()
show()
```
