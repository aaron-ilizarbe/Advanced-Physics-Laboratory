Para la excitación puntual de la barra se tiene un tren de onda descrito por: $10^{4}sen(10\pi t)$

```python
def seno1(x):
    return (10e4)*sin(10*pi*x)
x1 = linspace(0, 100, 10000)

plot(x1, seno1(x1))
xlim(6,8)
title("Vibración de la barra de acero")
xlabel("Tiempo (s)")
ylabel("Tren de onda")
show()
```

Para la excitación resonante de la barra se tiene un tren de onda que recubre las ondas en la FIG (3) descrito por: $2\times 10^{4}sen(5\pi t)$

```python
def seno2(x):
    return (2*(1e4))*sin(5*pi*x)
x2 = linspace(0, 100, 10000)
plot(x2, seno2(x2))
xlim(0,1)
ylim(-40000, 40000)
title("Vibración resonante de la barra de acero")
xlabel("Tiempo (s)")
ylabel("Tren de onda")
show()
```

Para la excitación resonante de la barra se tiene un tren de onda que recubre las ondas en la FIG (3) descrito por: $2\times 10^{4}sen(5\pi t + \pi)$ la cual recubre el tren de onda desfasado $\pi$.

```python
def seno3(x):
    return (2*(1e4))*sin(5*pi*x + pi)
x3 = linspace(0, 100, 10000)
plot(x3, seno3(x3))
xlim(0,1)
ylim(-40000, 40000)
title("Vibración resonante de la barra de acero")
xlabel("Tiempo (s)")
ylabel("Tren de onda")
show()
```

El tren de onda que se muestra en la FIG (3), se puede expresar como un producto de cosenos: $2\times 10^{4}cos(4.8\pi t)cos(52.8\pi t)$

```python
def cos4(x):
    return (20000)*cos(4.8*pi*x)*cos(52.8*pi*x)
x4 = linspace(0, 10, 10000)
plot(x4, cos4(x4))
xlim(0,1)
ylim(-40000, 40000)
title("Vibración resonante de la barra de acero")
xlabel("Tiempo (s)")
ylabel("Tren de onda")
show()
```
